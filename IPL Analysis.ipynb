# 3.Winning Toss - An advantage?

```python
df2.head()
```

```python
matches_per_season=df2.groupby('season')['season'].count()
matches_per_season
```

```python
import numpy as np
seasons=df2['season'].unique()
teams=df2['team1'].unique()
seasons=np.sort(seasons)
perc_win=[]
df_toss_match_winner=df2[df2['toss_winner']==df2['winner']]
df_toss_match_winner_season=df_toss_match_winner.groupby('season')['season'].count()
df_toss_match_winner_team=df_toss_match_winner.groupby('winner')['winner'].count()

for i in seasons:
    perc_win.append((df_toss_match_winner_season[i]/matches_per_season[i])*100)
```

```python
rects=plt.bar(seasons,perc_win)
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2., 1.05*height, str(int(height))+"%",ha='center', va='bottom',color='black',fontweight='bold')
plt.xticks(seasons,rotation="vertical")
plt.grid()
plt.ylim(0,80)
plt.xlabel("Seasons")
plt.ylabel("Percentage of wins")
plt.show()
```

##### From this we can observe the in the recent years there is slight advantage for winning the toss, but overall we see that its not a huge advantage

# 5. DL Method 

##### There has been a lot of speculation over the years that the Duck-Worth-Lewis method favours the team batting first. Lets check if its true

```python
df2.head()
```

```python
newdf=df2[df2["dl_applied"]!=0]
```

```python
newdf.reset_index(inplace=True)
```

```python
# Which team is batting first??not avaialable in table
```

```python
def set_batting_first(team1, team2, toss_winner, toss_decision):
    if toss_decision == 'bat':
        return toss_winner
    else:
        if team1 == toss_winner:
            return team2
        else:
            return team1
```

```python
newdf["batting first"]="1"
```

```python

for i in range(0,19):
    a=[]
    for j in range(5,9):
        a.append(newdf.iloc[i][j])
    x=set_batting_first(a[0],a[1],a[2],a[3])
    newdf.at[i, 'batting first'] = x
```

```python
newdf
```

```python
DL_win_bat_first=newdf[newdf["batting first"]==newdf["winner"]]
DL_win_bat_second=newdf[newdf["batting first"]!=newdf["winner"]]
```

```python
chasing=DL_win_bat_second.shape[0]
defending=DL_win_bat_first.shape[0]

```

```python
fig,ax=plt.subplots()
ax.pie([chasing,defending],
        labels=["batting first","batting second"],
        explode=(0.1,0.01),
        autopct="%.2f%%",
        shadow=True,
        startangle=90,
        textprops=dict(color="black"),
        colors=("cyan","orange"))

ax.set_title("WIN % when DL method is applied")
plt.show()
```
